## First Look at the Data & Visual Inspection

- First we take a look at the data in the different categorical columns to identify any invalid entries
- Then we convert the date columns to a datetime format to facilitate plotting each of the concerned variables against time and for later use.
- Finally, we plot the data.

```python
# use the .unique() to identify value types in the differet columns and identify obviously wrong categories or inputs

# make a list of the column headers with categorical values that can be visually inspected
columns = ["morning_s","afternoon_s","afternoon_r","morning_r","wind_e","wind_t"]

# define a function to make a list of all unique values in the specified columns.
def get_unique_categories(x,y):
    data = []
    try:
        for i in y:
            idata = [i]
            idata.append(x[i].unique())
            data.append(idata)
    except:
        print("bad data")
        pass
    return data

# call the function and print the unique values
a = get_unique_categories(mypd,columns)
print(a)
```

As we can see from the above the categorical columns (snow, rain, wind) do not have any unexpected values.

## Search for physically impossible values

We can also search computationally to flag values. The first type of these (which are categorically wrong) are physically impossible, here the following:
* temperature <0 K
* pressure <0 hPa)

```python
# returns values in a passed group of df columns that are lower than a specified value.
def check_lower_values(x,a):
    data = []
    for i in x:
        newdata = [i]
        for j in mypd[i]:
            if j < a:
                newdata.append(j)
        if newdata == [i]:
            newdata.append("none")
            data.append(newdata)
        else:
            data.append(newdata)
    return data

# assign columns with same cutoff value and call function
columns = ["temp_sunrise","temp_sunset"]
a = check_lower_values(columns,-270)

# assign columns with same cutoff value and call function
columns = ["p_sunrise","p_sunset"]
b = check_lower_values(columns,-270)

print(F"The impossible lower values are {a} and {b}")
```

## Physically implausible values
Once the impossible values are identifed we move on to the implausible values such as the following:
* temperatures outside -80 and +60 °C
* pressure outside 500 to 1100 hPa)

```python
#returns elements in column greater than a specified value
def check_upper_values(x,a):
    data = []
    for i in x:
        newdata = [i]
        for j in mypd[i]:
            if j > a:
                newdata.append(j)
        if newdata == [i]:
            newdata.append("none")
            data.append(newdata)
        else:
            data.append(newdata)
    return data

columns = ["temp_sunrise","temp_sunset"]
a = check_lower_values(columns,-80)
b = check_upper_values(columns,60)
wrongtemps = ("unlikely temperatures",[a,b])

columns = ["p_sunrise","p_sunset"]
a = check_lower_values(columns,500)
b = check_upper_values(columns,1100)
wrongpress = ("unlikely temperatures",[a,b])
print(wrongtemps)
print(wrongpress)
```

Note the two values visually inspected as being incorrect are now identified.

## Statistical tests
These tests make use of the full length of a series: for instance, values that exceed ±4 standard deviations can be flagged for further inspection.

```python
# function takes x = df, y = list of columns in df to analyse, returns list of values that exceed 4 standard deviations in their respective series.
def check_std(x,y):
    data = []
    # set test parameter to 4
    c = 4
    try:
        # loop through columns
        for j in y:
            # get standard deviation of values in x
            a = np.std(x[j])
            # get mean of values in x
            b = np.mean(x[j])
            # loop through the elements of each column
            for i in x[j]:
                # test occurs here
                if (i-b)/a > c:
                    # if true append name of column and failed test value to data list
                    datai = [j,i]
                    data.append(datai)
    except:
        print("its a string")
        data.append("not valid")
    return data

#list of columns to pass to check_std function
columns = ["p_sunrise", "p_sunset", "temp_sunrise","temp_sunset"]

more_than_four_stds = check_std(mypd,columns)
print(more_than_four_stds)
```
