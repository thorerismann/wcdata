## Search for physically impossible values

After examining the data, we search for different values that may indicate the data is wrong. The first type of these (which are categorically wrong) are physically impossible, such as the following:
* negative precipitation
* humidity >100%
* temperature <0 K
* pressure <0 hPa)

```python
# Check for temperatures below 0 K
negtemp_sunrise = mypd[mypd["tempe"] < -270]
negtemp_sunset = mypd[mypd["tempt"] < -270]
if negtemp_sunrise.empty & negtemp_sunset.empty:
    print("no negative temp values")
else:
    print("Some wrong  C values!")

# check for pressure < 0 hPa
negpressure_morning = mypd[mypd["mme"] < 0]
negpressure_afternoon = mypd[mypd["mmt"] < 0]
if negpressure_morning.empty & negpressure_afternoon.empty:
    print("no negative pressure values")
else:
    print("Some wrong hPa values!")

# Check for negative precipitation
negrain_morning = mypd[mypd["morning_r"] < 0]
negrain_afternoon = mypd[mypd["afternoon_r"] < 0]
negsnow_morning = mypd[mypd["morning_s"] < 0]
negsnow_afternoon = mypd[mypd["afternoon_s"] < 0]

if negrain_morning.empty & negrain_afternoon.empty & negsnow_morning.empty & negsnow_afternoon.empty:
    print("no negative precipitation values")
else:
    print("Some wrong precipitation values!")
```

## The second step
Next, search for physically implausible values such as the following:
* temperatures outside -80 and +60 C
* pressure outside 500 to 1100 hPa)

```python
# Check for Temperatures <-80 C and >60 C
ltm = mypd[mypd["tempe"] < -80]
htm = mypd[mypd["tempe"] > 60]
lta= mypd[mypd["tempt"] < -80]
hta= mypd[mypd["tempt"] > 60]

#concatonate implausible values together
badtemp = pd.concat([ltm,htm,lta,hta], sort = False)

# check if dataframe holding implausible values is empty
if badtemp.empty:
    print("no very unlikely temp values!")
else:
    print("At least one very unlikely  C values!")

# Same steps for to check for pressure < 500 hPa and > 1100 hPa
lpm = mypd[mypd["mme"] < 500]
hpm = mypd[mypd["mme"] > 1100]
lpa = mypd[mypd["mmt"] < 500]
hpa = mypd[mypd["mmt"] > 1100]
badpress= pd.concat([lpm, hpm, lpa, hpa], sort = False)

if badpress.empty:
    print("no very unlikely pressure values")
else:
    print("At least one very unlikely hPa value(s)!")
```

## The third step
Make use of the full length of a series. For instance, values that exceed 4 standard deviations are flagged for further inspection.

```python
mypd.columns
```

```python
# Check each value in the temperature,pressure and precipitation columns

# check_std tests each value in pd.series x against a preset number of standard deviations.
def check_std(x):
    data = []
    # get standard deviation of values in column
    a = np.std(x)
    # get mean of values in column
    b = np.mean(x)
    # set test parameter to 4
    c = 4
    try:
        for i in x:
            # test occurs here
            if (i-b)/a > c:
                data.append(i)
    except:
        print("its a string")
        data.append("not valid")
    return data

#store the results of check_std for each relevant column
sunset_pressure= check_std(mypd.mmt)
sunrise_pressure= check_std(mypd.mme)
temp_morning = check_std(mypd.tempe)
temp_evening = check_std(mypd.tempt)
```
